```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

  QUESTION 1
  

```python

titanic_data = pd.read_csv("titanic.csv")

# Task 1: Identify which deck had the highest survival rate
titanic_data['Deck'] = titanic_data['Cabin'].str[0]  # Extract deck information
deck_survival_rates = titanic_data.groupby('Deck')['Survived'].mean().sort_values(ascending=False)
print("Deck Survival Rates:")
print(deck_survival_rates)
```

QUESTION 2

```python
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1

plt.figure(figsize=(8, 6))
sns.histplot(titanic_data['FamilySize'], kde=False, bins=10, color='skyblue')
plt.title("Distribution of Family Sizes", fontsize=14)
plt.xlabel("Family Size", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

QUESTION 3

```python
third_class = titanic_data[titanic_data['Pclass'] == 3]
third_class_survival = third_class.groupby('Sex')['Survived'].mean()
print("\nSurvival Rates for Third Class Passengers:")
print(third_class_survival)
```

QUESTION 5


```python
median_fare = titanic_data['Fare'].median()
above_median_survivors = titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Fare'] > median_fare)]
percentage_above_median = (len(above_median_survivors) / len(titanic_data[titanic_data['Survived'] == 1])) * 100
print(f"\nPercentage of Survivors Who Paid Above the Median Fare: {percentage_above_median:.2f}%")
```

QUESTION 6

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=titanic_data, x='Pclass', y='Fare', hue='Survived', palette='viridis')
plt.title("Fare vs. Class Colored by Survival Status", fontsize=14)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.legend(title="Survived", loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
